```python
from flask import Flask, request, jsonify
from __future__ import print_function
import requests
import json
import cv2
import jsonpickle
```

```python
addr = 'https://reconhecimentofacialtelegram.herokuapp.com/reconhecimentofacial'

# prepare headers for http request
content_type = 'image/jpeg'
headers = {'content-type': content_type}

img = cv2.imread('trump_teste.jpg')
# encode image as jpeg
_, img_encoded = cv2.imencode('.jpg', img)
# send http request with image and receive response
response = requests.post(addr, data=img_encoded.tostring(), headers=headers)
# decode response

#print(response.text)
if not response.text is None:
    frame = jsonpickle.decode( response.text ) 
    #img = frame.tobytes()
    img=cv2.imdecode(frame,cv2.IMREAD_COLOR)

    if not frame is None:
        cv2.imshow("Output", img)
        cv2.waitKey(0)
        cv2.destroyAllWindows()

# expected output: {u'message': u'image received. size=124x124'}
```
